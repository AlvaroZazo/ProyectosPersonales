```python
import torch
from torch.autograd import Variable
import numpy as np
import matplotlib.pyplot as plt
```

```python
valores = torch.from_numpy(np.array([a**2 for a in np.arange(-32., 33.)]).astype("float32"))
```

```python
plt.plot(valores)
plt.show()
```

```python
valoresVar = Variable(valores, requires_grad=True)
valoresMAX = torch.max(valoresVar * valoresVar)
valoresMAX.backward()
```

```python
valoresVar.grad
```

```python
plt.plot(valoresVar.data.numpy(), label="valores")
plt.plot(valoresVar.grad, label="gradientes")
plt.legend()
plt.show()
```

```python
predicted_outputVar = Variable(valores, requires_grad=True)
expected_output = torch.zeros(len(valoresVar))
loss = (predicted_outputVar - expected_output).pow(2).sum()
loss.backward()
```

```python
predicted_outputVar.grad
```

```python
plt.plot(predicted_outputVar.data.numpy(), label="valores")
plt.plot(predicted_outputVar.grad, label="gradientes")
plt.legend()
plt.show()
```
